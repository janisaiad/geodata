```python
import os
from matplotlib import pyplot as plt
from matplotlib.image import imread
```

```python

import numpy as np

def show(img, cmap='gray', ax=None, title=None):
    if ax is None:
        fig, ax = plt.subplots()
    ax.imshow(img, cmap=cmap)
    if title:
        ax.set_title(title)
    ax.axis('off')
    # Show image dimension and tensor
    print(f"Image shape: {img.shape}")
    print("Tensor (numpy array):")
    # Print a small part if it's a large array
    arr = np.array(img)
    if arr.size > 100:
        print(arr.ravel()[:100], "...")  # print first 100 elements
    else:
        print(arr)
    plt.show()

data_dir = '../data/pixelart/images'
image_files = [f for f in os.listdir(data_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif'))]

images = []
for fname in sorted(image_files):
    path = os.path.join(data_dir, fname)
    img = imread(path)
    images.append((fname, img))

for fname, img in images:
    print(f"======= {fname} =======")
    show(img, title=fname)
```
